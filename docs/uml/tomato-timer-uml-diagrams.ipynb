## 1. 核心计时功能用户故事

### US001: 启动番茄计时器
**作为** 一个需要专注工作的用户  
**我希望** 能够点击开始按钮启动25分钟的工作计时器  
**以便于** 开始一个标准的番茄工作时段  

**验收标准：**
- ✅ 点击开始按钮后，计时器从25:00开始倒计时
- ✅ 计时器状态显示为"工作中"
- ✅ 开始按钮变为暂停按钮
- ✅ 圆形进度条开始显示进度

### US002: 暂停和恢复计时器
**作为** 一个正在工作的用户  
**我希望** 能够临时暂停计时器并稍后恢复  
**以便于** 处理紧急情况而不丢失当前进度  

**验收标准：**
- ✅ 点击暂停按钮后，计时器停止倒计时
- ✅ 暂停按钮变为继续按钮
- ✅ 点击继续按钮后，计时器从暂停位置继续倒计时
- ✅ 进度条状态保持不变

### US003: 停止当前计时
**作为** 一个用户  
**我希望** 能够完全停止当前的计时会话  
**以便于** 重新开始或结束当前任务  

**验收标准：**
- ✅ 点击停止按钮后，计时器重置为初始状态
- ✅ 时间显示回到25:00
- ✅ 按钮状态重置为开始按钮
- ✅ 进度条重置为0%

### US004: 自动会话转换
**作为** 一个番茄工作法实践者  
**我希望** 工作时段结束后自动转换为休息时段  
**以便于** 遵循标准的番茄工作法节奏  

**验收标准：**
- ✅ 25分钟工作时间结束后，自动切换为5分钟短休息
- ✅ 完成4个工作时段后，自动切换为15分钟长休息
- ✅ 会话类型正确显示（工作/短休息/长休息）
- ✅ 播放完成提示音
